```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = 'factbook.csv'
info_df = pd.read_csv(filename, delimiter=';')
cov_df = pd.read_csv('finished_datasets/country.csv')
```

```python
info_df.head()
```

## Matching countries

```python
setcov = set(cov_df.country.unique())
setinfo = set(info_df.Country.unique())
setcov - setinfo
```

Fixing naming inconsistencies

```python
info_df = info_df.replace('Bahamas The', 'The Bahamas')
info_df = info_df.replace('Czech Republic', 'Czechia')
info_df = info_df.replace("Cote d'Ivoire", "Côte d'Ivoire")
```

```python
country_mask = info_df.Country.isin(cov_df.country.unique())
info_df = info_df[country_mask]
```

```python
info_df.isna().sum()
```

```python
info_df.to_csv(r'finished_datasets/country_info.csv', index = False)
```
